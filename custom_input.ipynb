```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

'''    Class giao diện của tool    '''

class GUI:
    def __init__(self, root):
        self.root = root

        # Điều chỉnh kích cỡ và thuộc tính giao diện
        self.root.title("AutoFill Google Forms")
        self.root.geometry("600x400")
        self.root.resizable(False, False)
        self.root.config(bg="#d3d3d3")

        # Tạo notebook để dễ dàng chuyển trang, notebook gồm 4 trang
        self.notebook = ttk.Notebook(self.root)
        self.notebook.pack(fill='both', expand=True)
        
        self.frame1 = ttk.Frame(self.notebook)
        self.frame2 = ttk.Frame(self.notebook)
        self.frame3 = ttk.Frame(self.notebook)
        self.frame4 = ttk.Frame(self.notebook)
        
        # scrollable_frame = ScrollableFrame(self.frame2) # Tạo thanh cuộn cho trang 2
        # scrollable_frame.pack(fill="both", expand=True)

        self.notebook.add(self.frame1, text="Form Details")
        self.notebook.add(self.frame2, text="Page Details")
        self.notebook.add(self.frame3, text="Prefill Links")
        self.notebook.add(self.frame4, text="Filling Progress")

        # self.current_page = 0
        # self.pages = []
        # self.field_count = 0

        self.init_tab1() # Khởi tạo tab 1
        # self.init_tab3()
        # self.init_tab4()

        # # Loading Screen
        # self.loading_screen = None
        # self.progress_var = tk.DoubleVar()

    def init_tab1(self):
        # Số form cần điền
        ttk.Label(self.frame1, text="Number of forms:").grid(column=0, row=0, padx=10, pady=10, sticky='W')
        self.num_forms_entry = ttk.Entry(self.frame1)
        self.num_forms_entry.grid(column=1, row=0, padx=10, pady=10, sticky='W')

        # Số trang của form
        ttk.Label(self.frame1, text="Number of pages per form:").grid(column=0, row=1, padx=10, pady=10, sticky='W')
        self.num_pages_entry = ttk.Entry(self.frame1)
        self.num_pages_entry.grid(column=1, row=1, padx=10, pady=10, sticky='W')

        # Nút chuyển tiếp
        self.tab1_Next_button = ttk.Button(self.frame1, text="Next", command=self.init_tab2)
        self.tab1_Next_button.grid(column=2, row=2, padx=10, pady=10, sticky='W')

    # Chuyển sang tab 2 sau khi ấn nút
    def init_tab2(self):
        self.notebook.select(self.frame2)
        self.num_forms_entry.config(state='disabled') # ngăn chỉnh sửa giá trị ở tab đã điền
        self.num_pages_entry.config(state='disabled')
        self.tab1_Next_button.config(state='disabled')

        # Lấy các giá trị vừa điền ở tab1
        try:
            self.num_forms = int(self.num_forms_entry.get())
            self.num_pages = int(self.num_pages_entry.get())
        except ValueError:
            messagebox.showerror("Input Error", "Please enter a valid number of pages.")

        self.field_count = [0] * self.num_pages
        # for widget in self.frame2.winfo_children():
        #     widget.destroy()
        for i in range(self.num_pages):
            ttk.Label(self.frame2, text=f"Page {i + 1}").pack(padx=10, pady=10)
            self.add_field_button = ttk.Button(self.frame2, text="Add Field", command=lambda p=i: self.add_field(p))
            self.add_field_button.pack(padx=10, pady=10)
            
        self.current_page = 0
        self.show_page(self.current_page)


    def add_field(self, page_index):
        page_frame = self.pages[page_index]
        fields_frame = ttk.Frame(page_frame)
        fields_frame.pack(pady=5, fill='x')

        ttk.Label(fields_frame, text=f"Field {self.field_count + 1} XPath:").grid(column=0, row=self.field_count, padx=5, pady=5, sticky='W')
        ttk.Entry(fields_frame).grid(column=1, row=self.field_count, padx=5, pady=5, sticky='W')

        ttk.Label(fields_frame, text="Data:").grid(column=2, row=self.field_count, padx=5, pady=5, sticky='W')
        ttk.Combobox(fields_frame, values=["Manual", "Auto Name", "Auto Phone", "Auto Email"]).grid(column=3, row=self.field_count, padx=5, pady=5, sticky='W')

        self.field_count += 1

    def show_page(self, page_index):
        for i, page in enumerate(self.pages):
            if i == page_index:
                page.pack(fill='x', expand=True)
            else:
                page.pack_forget()

    def init_tab3(self):
        ttk.Label(self.frame3, text="Enter prefill links (one per line):").grid(column=0, row=0, padx=10, pady=10, sticky='W')
        self.prefill_links_text = tk.Text(self.frame3, height=10)
        self.prefill_links_text.grid(column=0, row=1, padx=10, pady=10, sticky='W')

    def init_tab4(self):
        ttk.Button(self.frame4, text="Start", command=self.start_filling).grid(column=0, row=0, padx=10, pady=10, sticky='W')
        ttk.Button(self.frame4, text="Pause", command=self.pause_filling).grid(column=1, row=0, padx=10, pady=10, sticky='W')
        ttk.Button(self.frame4, text="Continue", command=self.continue_filling).grid(column=2, row=0, padx=10, pady=10, sticky='W')

        # Filling mode
        ttk.Label(self.frame4, text="Filling Mode:").grid(column=0, row=1, padx=10, pady=10, sticky='W')
        self.filling_mode = tk.StringVar(value="Random")
        ttk.Radiobutton(self.frame4, text="Random", variable=self.filling_mode, value="Random").grid(column=1, row=1, padx=5, pady=5, sticky='W')
        ttk.Radiobutton(self.frame4, text="Fair", variable=self.filling_mode, value="Fair").grid(column=2, row=1, padx=5, pady=5, sticky='W')

    def start_filling(self):
        messagebox.showinfo("Info", "Starting to fill forms...")

    def pause_filling(self):
        messagebox.showinfo("Info", "Pausing form filling...")

    def continue_filling(self):
        messagebox.showinfo("Info", "Continuing form filling...")

    def show_loading_screen(self):
        self.loading_screen = tk.Toplevel(self.root)
        self.loading_screen.title("Loading")
        # self.loading_screen.geometry("300x100")
        ttk.Label(self.loading_screen, text="Filling forms...").pack(pady=10)
        ttk.Progressbar(self.loading_screen, mode='determinate', variable=self.progress_var).pack(pady=10)

    def update_progress(self, value):
        self.progress_var.set(value)

if __name__ == "__main__":
    root = tk.Tk()
    app = GUI(root)
    root.mainloop()
```
